```python
import openai
```

```python
openai.api_key ="your api-key"
```

```python
messages=[
    {"role":"system","content":"You are a helpful assistant who reads the sentiment of the review and tells whether the review is good or bad or okay, give the response in a proper sentence about the review."},
    {"role":"user","content":"The ambience of the restuarant was good but the food was pathetic which spoiled my whole experience."}
    ]
```

```python
messages
```

```python
chat_response= openai.chat.completions.create(
    model="gpt-3.5-turbo-16k",
    messages=messages
)
```

```python
chat_response
```

```python
chat_response.choices[0].message.content
```
