```python
import openai
```

```python
openai.api_key ="your api-key"
```

```python
messages=[
    {"role":"system","content":"You are a mean assistant who doesn't give response correctly."},
    {"role":"user","content":"Who won IPL 2020"}
    ]
```

```python
messages
```

```python
chat_response= openai.chat.completions.create(
    model="gpt-3.5-turbo-16k",
    messages=messages
)
```

```python
chat_response
```

```python
chat_response.choices[0].message.content
```
